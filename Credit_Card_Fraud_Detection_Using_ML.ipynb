Import of the Dependencies

```python
#import of libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#dataset loading to a pandas DataFrame
card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#the first five data of the dataset
card_data.head()
```

```python
#observing the final five data of the dataset. It is important because it gives an overall idea about the dataset
card_data.tail()
```

```python
#more information about dataset
card_data.info()
```

```python
#In case any missing values in any column
card_data.isnull().sum
```

```python
#distribution of legit(--> 0) & fradulant transaction(--> 1)
card_data['Class'].value_counts()
```

Data is highly unbalanced

```python
#seperation of data for analysis
legit = card_data[card_data.Class == 0]
fraud = card_data[card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#iportant statistical measure of legit data
legit.Amount.describe()
```

```python
#iportant statistical measure of fraud data
fraud.Amount.describe()
```

```python
#comparison of the values of both transaction
card_data.groupby('Class').mean()
```

Under - Sampling

Building a simple dataset containing similar distribution of normal and fraudulant transaction

Number of fraudulant transaction is 492

```python
legit_sample = legit.sample(n=401)
```

Concatenation of two data frames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting data into features and targets

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state= 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Create a scaler
scaler = StandardScaler()

# Fit and transform the scaler on the training data
X_train_scaled = scaler.fit_transform(X_train)

# Create and train the logistic regression model with increased iterations on the scaled data
model = LogisticRegression(max_iter=1000)
model.fit(X_train_scaled, Y_train)
```

```python
#training logisticRegression model with training data
model.fit(X_train_scaled, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
